# Trade-off between Partitioning ratio and Discrepancy

In this notebook we explore the trade-off between the partitioning ratio $\alpha$ and the discrepancy $\beta$, as detailed in Section 5 of the paper.

```python
import matplotlib.pyplot as plt
import numpy as np
import tikzplotlib

def f(p):
    def f_p(alpha):
        return (alpha*(2-alpha))**(1/p)

    return f_p

def g(p):
    def g_p(alpha):
        return alpha**(1/p)
    return g_p


def gcomp(p):
    def g_p(beta):
        return beta**p
    return g_p


plt.figure(figsize=(6,6))
plt.plot(f(2)(np.linspace(0,1,100)),(np.linspace(0,1,100)), label="MaxNash")
plt.plot(f(1)(np.linspace(0,1,100)),(np.linspace(0,1,100)), label="MaxSum")
plt.plot(g(2)(np.linspace(0,1,100)),(np.linspace(0,1,100)), label="Half0-Nash")
plt.plot(g(1)(np.linspace(0,1,100)),(np.linspace(0,1,100)), label="HalfNeg-Nash")

# areas 
plt.fill_between(f(2)(np.linspace(0,1,100)), 0, (np.linspace(0,1,100)), alpha=0.3)
plt.fill_between(f(1)(np.linspace(0,1,100)), 0, (np.linspace(0,1,100)), alpha=0.3)
plt.fill_between(g(2)(np.linspace(0,1,100)), 0, (np.linspace(0,1,100)), alpha=0.3)
plt.fill_between(g(1)(np.linspace(0,1,100)), 0, (np.linspace(0,1,100)), alpha=0.3)

plt.legend()
plt.xlabel("Local Polarization beta(x,y)")
plt.ylabel("Global Polarization alpha(a,b)")
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("Limit ab/xy (ab win on top left)")
tikzplotlib.save("figures/rules.tex")
plt.show()
```
